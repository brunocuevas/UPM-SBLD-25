- Once you've located the UniProt entry and obtained its UniProt ID, use the ChEMBL API to find the corresponding ChEMBL identifier for that enzyme. This will involve sending a request to the ChEMBL API and extracting the relevant information from the response.


## Exercise 2: Find the Most Potent Compound

## Exercise 3: Similarity Search in Chembl

- Experiment with different similarity thresholds to see how the results change.
- Note down a few of the similar compounds retrieved.

## Exercise 4: Similarity Search in PubChem

- Now, perform a 2D similarity search using the PubChem API, using the SMILES of the most potent compound.
- Compare the results obtained from Chembl and PubChem searches.
- Note down any differences or similarities in the retrieved compounds.
